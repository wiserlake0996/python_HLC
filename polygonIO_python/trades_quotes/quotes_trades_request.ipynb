```python
import datetime
import pandas as pd
import pytz
eastern = 'US/Eastern'

from polygon import RESTClient
```

```python
key = "qRFktbNL6A_WzaFhqR80IllApsmjPY6a"

def ts_to_datetime(ts) -> str:
    return datetime.datetime.fromtimestamp(ts / 1000.0).strftime('%Y-%m-%d %H:%M')

def get_list_of_Exchanges():

    with RESTClient(key) as client:
        respex = client.stocks_equities_exchanges()

        stock_exchanges = respex.exchange
        exchange_list_dict = {}

        for i in stock_exchanges:
            the_dict = {}
            the_dict[i.i_d_of_the_exchange] = i.name
            #exchange_list_dict.append(the_dict)
            exchange_list_dict[i.i_d_of_the_exchange] = i.name
        return exchange_list_dict

def get_list_of_trade_conditions():

    with RESTClient(key) as client:
        respex = client.stocks_equities_condition_mappings()

        stock_exchanges = respex.exchange
        exchange_list_dict = {}

        for i in stock_exchanges:
            the_dict = {}
            the_dict[i.i_d_of_the_exchange] = i.name
            #exchange_list_dict.append(the_dict)
            exchange_list_dict[i.i_d_of_the_exchange] = i.name
        return exchange_list_dict
```

```python

def update_condition_labels(keyFunction, values):
    k =  list(keyFunction[str(v)] for v in list(values))
    return ", ".join(k)

```

```python
def get_quotes(ticker, date, size):
        
    with RESTClient(key) as client:
        resp = client.historic_n___bbo_quotes_v2(ticker=str(ticker).upper(), date=date, limit=size, reverse=False)
        return resp.__dict__
    
def get_trades(ticker, date, size):
        
    with RESTClient(key) as client:
        resp = client.historic_trades_v2(ticker=str(ticker).upper(), date=date, limit=size, reverse=False)
        return resp.__dict__
```

# To make a quote request, inputs for get_quotes func

###### get_quotes([ticker], [date], [limit])
###### max limit is 50000
    

```python
quotes = get_quotes("AAPL", "2020-10-14", 300)
```

```python
quotes_df = pd.DataFrame(quotes['results']).rename(columns=quotes_map)
```

```python
quotes_df = quotes_df.replace({'bid_exchange':exchange_list_dict, 'ask_exchange':exchange_list_dict})
quotes_df['conditions'] = quotes_df['conditions'].apply(lambda x: update_condition_labels(quote_conditions, x))
```

```python
quotes_df['sip_timestamp'] =pd.to_datetime(quotes_df['sip_timestamp']).dt.tz_localize('utc').dt.tz_convert(eastern)
quotes_df['participant_timestamp'] =pd.to_datetime(quotes_df['participant_timestamp']).dt.tz_localize('utc').dt.tz_convert(eastern)
```
